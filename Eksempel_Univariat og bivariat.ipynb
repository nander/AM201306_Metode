```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('Data/Bivariate data-ForExamples.csv')

# Display the first few rows of the DataFrame
df.head()
```

Display the summary statistics for the loaded data.

```python
df.describe()
```

Generate histograms for the numerical variables in the DataFrame.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Set up the figure size to accommodate multiple plots
plt.figure(figsize=(15, 10))

# Loop through each numerical column and create a histogram
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 4, i + 1) # Adjust subplot grid as needed
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Create a scatterplot between each pair

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['SD_x'], y=df['SD_y'])
plt.title('Scatter Plot of SD_x vs SD_y')
plt.xlabel('SD_x')
plt.ylabel('SD_y')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
