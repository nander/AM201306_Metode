# Kodebok for Eiendomsdata

Denne kodeboken beskriver kolonner i datasettet over eiendomssalg i Norge.

## Variabler

### pris

Beskrivelse: Salgspris for eiendommen.

Enhet: Norske kroner (NOK).

### pris_kvm

Beskrivelse: Salgspris per kvadratmeter.

Enhet: Norske kroner per kvadratmeter (NOK/m²).

### prisant

Beskrivelse: Opprinnelig prisantydning for eiendommen.

Enhet: Norske kroner (NOK).

### verditakst

Beskrivelse: Verditakst for eiendommen.

Enhet: Norske kroner (NOK).

### BOAProm

Beskrivelse: Bruksareal (BOA) for eiendommen.

Enhet: Kvadratmeter (m²).

### energi

Beskrivelse: Energimerking av eiendommen.

Enhet: Energiklasse (A til G).

boligtype

Beskrivelse: Type bolig (f.eks. Enebolig, Rekkehus, Leilighet, Tomannsbolig).

Enhet: Kategorisk.

vest

Beskrivelse: Indikator for beliggenhet i vestlige bydeler.

Enhet: Binær (0 = ikke vest, 1 = vest).

dummyGrünneløkkaSagene

Beskrivelse: Indikator for beliggenhet i bydel Grünneløkka/Sagene.

Enhet: Binær (0 = nei, 1 = ja).

### dummyFrogner

Beskrivelse: Indikator for beliggenhet i bydel Frogner.

Enhet: Binær (0 = nei, 1 = ja).

### dummySt.Hanshaugen

Beskrivelse: Indikator for beliggenhet i bydel St. Hanshaugen.

Enhet: Binær (0 = nei, 1 = ja).

### dummyGamleOslo

Beskrivelse: Indikator for beliggenhet i bydel Gamle Oslo.

Enhet: Binær (0 = nei, 1 = ja).

### solgt

Beskrivelse: Årstall for når eiendommen ble solgt.

Enhet: År (f.eks. 2014).

### byggeår

Beskrivelse: Byggeår for eiendommen.

Enhet: År (f.eks. 1919).

### alder

Beskrivelse: Alder på eiendommen ved salgsåret.

Enhet: År.
